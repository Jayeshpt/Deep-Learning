```python
import numpy as np
```

```python
# m_current = slope , intercept = b
import matplotlib.pyplot as plt

def gradient_descent(x,y):
    m_curr = b_curr = 0
    iterations = 1000
    n = len(x)
    learning_rate = 0.01
    cost_list = []  # Add an empty list to store the cost values
    
    for i in range(iterations):
        y_predicted = m_curr * x + b_curr   ### y=mx+c
        cost  = (1/n) * sum([val**2 for val in (y-y_predicted)])
        md = -(2/n) *sum(x*(y-y_predicted)) ### finding m derivative  
        bd = -(2/n) *sum(y-y_predicted)     ### b derivative
        m_curr = m_curr- learning_rate * md   ### Adjusting m current and  b current with learning rate
        b_curr = b_curr- learning_rate * bd
        cost_list.append(cost)  # Append the cost value to the list
        print('m {} || b {} || cost {} || iteration {}'.format(m_curr,b_curr,cost,i))
    
    # Plot the cost vs. iteration
    plt.plot(cost_list)
    plt.xlabel('Iteration')
    plt.ylabel('Cost')
    plt.show()

# Call the function with some data
x = np.array([1,2,3,4,5])
y =np.array([5,6,7,8,9])  
gradient_descent(x,y)
```
